```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# эмулятор динамики цен активов (рынка ценных бумаг, валют, металлов и т.д.)
def stock_gen(num, per, pri, med, std):
    mask = np.full((num, 1), pri / num)
    return np.hstack((mask, np.random.normal(med, std, (num, per)) * mask))
```

```python
# стартовый капитал
c = 1000
# период
p = 24
# медианный уровень роста активов
med = 0.01
# стандартное отклонение
std = 0.05

# количество измерений
count = 100

# список числа активов 
n_list = [1,2,3,5,10,30,50,100,200]

period = np.arange(1+p)
```

```python
fig = plt.figure(figsize=(10, 10), dpi=80)

for i, val in enumerate(n_list):
    ax = fig.add_subplot(3,3,i+1)
    for _ in range(count):
        stock_pr = stock_gen(val, p, c, med, std)
        result = np.cumsum(np.sum(stock_pr, axis=0))
        ax.plot(period, result)
        ax.set_title(f"n={val}")

plt.show()
```
